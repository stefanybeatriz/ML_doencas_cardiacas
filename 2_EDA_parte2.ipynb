#### <a id = "Análise_Bivariada_ou_Multivariada">Análise Bivariada ou Multivariada</a>

***Obtendo informações sobre o comportamento entre duas ou mais variáveis***

Assim como na análise univariada, a análise bivariada ou multivariada envolve tanto o cálculo de estatísticas resumidas quanto a produção de representações visuais que nos permite perceber o comportamento entre duas ou mais variáveis, isto é, se há algum tipo de dependência estatística entre elas, o que não implica necessariamente que mudanças em uma variável seja a causa de mudanças na outra. O tipo adequado de ferramenta a ser utilizada para os devidos fins dependerá da natureza dos dados, isso é, dos tipos de variáveis que estarão sendo analisadas (quantitativas ou qualitativas), da característica de suas distribuições e do objetivo da análise, onde a determinação correta dessas ferramentas garantirá um resultado mais fidedigno e confiável.

#### <a id = "Correlação">Correlação</a>

Quando estamos diante da necessidade de entendimento sobre a magnitude e a direção da **relação estatística** entre variáveis, sejam elas **lineares** ou **não lineares**, estamos diante do conceito estatístico de **correlação**.

<div class="alert alert-block alert-warning">
    <b><i>Importante:</i></b> Embora correlação seja informalmente entendida como sinônimo de dependência, e mesmo que se constate uma correlação considerável entre duas variáveis, isso não significa necessariamente - como já observado - que uma mudança em uma variável seja a causa de uma mudança na outra. Definitivamente, <b>correlação não implica causalidade</b>.
</div>

Existem várias maneiras de se entender tanto visualmente quanto quantitativamente a correlação entre variáveis. As mais comuns são o **gráfico de dispersão (*scatterplot*)** e os **coeficientes de correlação ou associação**.

#### <a id = "Gráfico_de_Dispersão">Gráfico de Dispersão</a>

O gráfico de dispersão (*scatterplot*) é a forma mais comum de se visualizar o comportamento entre duas ou mesmo mais variáveis quantitativas, plotando-as entre si via coordenadas cartesianas, onde o valor de uma variável determina a posição no eixo horizontal e o valor da outra variável determina a posição no eixo vertical. Assim, cada par de valor é representado por um ponto no gráfico.

O objetivo do gráfico de dispersão é permitir a identificação de diferentes tipos de relação entre as variáveis, como uma relação linear, uma relação não linear, se a relação é positiva, negativa ou mesmo nula, além de *outliers* e a consistência dos dados.

#### <a id = "Coeficiente_de_Correlação_de_Pearson">Coeficiente de Correlação de Pearson</a>

Antes de falarmos sobre o coeficiente de correlação de Pearson, precisamos entender a **covariância**, a qual uma medida de variabilidade conjunta entre duas variáveis. É uma medida de como as alterações em uma variável estão **linearmente** associadas às mudanças em uma outra variável.

Quando a covariância entre essas variáveis é positiva, os dados tendem a variar na mesma direção, isso é, enquanto os valores de uma variável aumentam, os valores da outra também aumentam. Quando a covariância entre essas variáveis é negativa, os dados tendem a variar em direções opostas, isso é, enquanto os valores de uma variável aumentam, os valores da outra diminuem.

> A covariância é a somatória dos produtos dos desvios da média das variáveis *X* e *Y* divido por *n* (populacional) ou *n-1* (amostral).

O **coeficiente de correlação de Pearson** é uma versão normalizada da covariância que demonstra a **magnitude e a direção da relação linear** entre duas variáveis, representada por valores entre **−1** e **+1**. Assim como a própria covariância, essa medida reflete apenas uma correlação linear de variáveis e ignora muitos outros tipos de relacionamentos.

> O coeficiente de correlação de Pearson é a covariância das duas variáveis dividida pelo produto de seus desvios padrão.

A interpretação dos valores do coeficiente de correlação de Pearson se dá da seguinte maneira:

|coeficiente|magnitude e direção|
|----------:|:------------------|
|+1|relação linear positiva perfeita|
|>0|relação linear positiva|
||*toda vez que x aumenta, y aumenta*|
|=0|relação linear nula|
||*x e y não dependem linearmente uma da outra*|
|<0|relação linear negativa|
|-1|relação linear negativa perfeita|
||*toda vez que x aumenta, y diminui*|

Quando as variáveis $X$ e $Y$ são independentes, o coeficiente de correlação de Pearson é $0$ (zero). Mas é importante salientar que o contrário não é verdadeiro porque esse coeficiente de correlação detecta apenas dependências lineares entre duas variáveis.

Como exemplo, sendo uma variável $X$ simetricamente distribuída em torno de $0$ (zero) e uma variável $Y$ que seja o resultado de $X^2$, temos que $Y$ é completamente determinada por $X$, de modo que sejam então perfeitamente dependentes. Fato é que, ao medir a correlação entre elas, o coeficiente resulta em $0$ (zero), indicando que as variáveis não são correlacionadas. Perceba o quanto esse coeficiente de correlação requer cuidado em sua interpretação, pois a equivalência entre não correlação e independência entre variáveis pressupõe que $X$ e $Y$ sejam conjuntamente normais.

```python
np.random.seed(0)

# randn() gera uma matriz de números aleatórios normalmente distribuídos com média 0 e desvio padrão 1
var_1 = np.random.randn(10000) # X
var_2 = var_1**2 # Y

plt.hist(var_1, bins=100, alpha=.67)
plt.hist(var_2, bins=100, alpha=.67)
plt.show()
```

```python
# plotando o gráfico de dispersão (scatterplot)
plt.scatter(var_1, var_2)
plt.show()
```

```python
var_1 = pd.Series(var_1)
var_2 = pd.Series(var_2)

# calculando o coeficiente de correlação de Pearson entre duas variáveis (valores nulos (NaN) são excluídos por default)
var_1.corr(other=var_2, method='pearson')
```

#### Casos especiais do coeficiente de correlação de Pearson

> *coeficiente de correlação de Pearson = coeficiente de correlação ponto bisserial*

- Quando as duas variáveis são naturalmente dicotômicas

#### <a id = "Coeficiente_de_Correlação_de_Spearman">Coeficiente de Correlação de Spearman</a>

O **coeficiente de correlação de Spearman** é especialmente útil quando a relação entre as variáveis não pode ser facilmente descrita por uma função linear, caso esse onde o coeficiente de correlação de Pearson falharia. Esse coeficiente também demonstra a magnitude e a direção da relação estatística entre duas variáveis, sendo capaz de capturar correlações **monotônicas**, sejam elas positivas ou negativas, de características não lineares. Além de não exigir a suposição de que haja relacionamento linear entre as variáveis, ele também não requer que as mesmas estejam distribuidas normalmente, podendo inclusive ser utilizado para medir o relacionamento entre **variáveis de característica ordinal**. Por essas características, esse coeficiente é uma medida de correlação não paramétrica a qual se baseia na **relação entre a ordem dos dados**.

Sua interpretação é a mesma que o do coeficiente de correlação de Pearson.

#### Codificando a variável *target* para um estado ordinal

```python
# definindo o mapeamento das categorias para valores numéricos
cat2num_multiclass = {'healthy': 0, 'sick-low': 1, 'sick-medium': 2, 'sick-high' : 3, 'sick-serious' : 4}

# aplicando o mapeamento
df_raw['Diagnosis Num (multiclass)'] = df_raw['Diagnosis (multiclass)'].map(cat2num_multiclass)

df_raw.head()
```

```python
df_raw['ST Depression (exercise-induced)'].plot.hist()
```

```python
var_1 = df_raw['ST Depression (exercise-induced)']
var_2 = df_raw['Diagnosis Num (multiclass)']

# calculating the correlation coefficient between two variables (excluding NaN values)
var_1.corr(var_2, method='spearman')
```

```python
# plotando o gráfico de dispersão (scatterplot)
df_raw.plot.scatter(x='ST Depression (exercise-induced)', y='Diagnosis Num (multiclass)')
```

#### <a id = "Associação">Associação</a>

A **associação** refere-se à qualquer tipo de relação estatística entre duas variáveis, não se limitando apenas à natureza linear ou monotônica dessas relações, incluindo então relações curvilíneas e complexas que não podem ser facilmente capturadas por um coeficiente de correlação como o de Pearson (relação linear) ou mesmo por um mais flexível como o de Spearman (não linear mas monótona).

Existem várias maneiras de se entender tanto visualmente quanto quantitativamente a assiciação entre variáveis. As mais comuns são a **tabela de contigência** (ou mesmo por um gráfico que melhor a represente) e o **coeficiente de associação Cramér's V**.

#### <a id = "Coeficiente_de_Associação_Cramér's_V">Coeficiente de Associação Cramér's V</a>

O **coeficiente de associação Cramér's V** é uma medida de associação estatística que quantifica a magnitude da associação entre duas variáveis categóricas, nos auxiliando a compreender a existência de uma relação estatisticamente significativa entre elas. Esse coeficiente é calculado como uma versão normalizada da estatística **qui-quadrado (*χ²*)** e, sendo assim, leva em consideração o número de categorias nas variáveis e o tamanho total da amostra (extraídos através da confecção de uma tabela de contingência). Esse coeficiente pode detectar associações não lineares, monotônicas e não monotônicas, sendo uma medida mais robusta para casos em que a relação entre as variáveis se apresente de modo complexo, a qual não pode ser adequadamente capturada pelos coeficientes de correlações como já citados.

O coeficiente de associação Cramér's V varia de $0$ a $1$, onde $0$ indica ausência total de associação e $1$ indica associação perfeita entre as variáveis. Sendo assim, valores mais próximos de $1$ representam uma associação mais forte, enquanto valores próximos de $0$ indicam uma associação mais fraca.

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.crosstab.html

```python
# Criando uma tabela de contingência
contingency_table = pd.crosstab(df_raw['Sex'], df_raw['Diagnosis (multiclass)'])
display(contingency_table)
```

https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.contingency.association.html

```python
from scipy.stats.contingency import association

# Calculando o coeficiente de associação Cramér's V
cramer_v = association(contingency_table)

print("Coeficiente de Associação Cramér's V:", cramer_v)
```

```python
contingency_table.plot.bar().legend(bbox_to_anchor=(1.33, 1.0))
```

```python
contingency_table.plot.bar(stacked=True).legend(bbox_to_anchor=(1.0, 1.0))
```

A utilização de **Gráficos de Barras** possibilita várias oportunidades de entendimentos sobre como as variáveis quantitativas (numéricas) e principalmente as qualitativas (categóricas) se relacionam em um determinado contexto.

***Exemplo:*** Visualizar a média de colesterol dos pacientes de acordo com as categorias feminino e masculino.

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.groupby.html

```python
df_raw.groupby('Sex')['Cholesterol (total)'].mean().plot.bar()
```

***Exemplo:*** Visualizar tanto a frequência absoluta quanto a frequência relativa das categorias feminino e masculino por resultado positivo ou negativo do teste de glicose.

```python
# 'normalize=False' returns the absolute frequency
df_raw.groupby('FBS Test')['Sex'].value_counts().plot.bar() # 'value_counts()' has 'normalize=False' by default
```

```python
# 'normalize=True' returns the relative frequency
df_raw.groupby('FBS Test')['Sex'].value_counts(normalize=True).plot.bar()
```
