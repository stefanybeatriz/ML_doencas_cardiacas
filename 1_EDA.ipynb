### <a id = "Motivação"><font color='#005b96'>Motivação</font></a>

- **Cenário Global**

Segundo a *World Health Organization*, as doenças cardiovasculares (DCVs) são a principal causa de morte no mundo todo, com cerca de 17,9 milhões por ano. Mais de quatro em cada cinco mortes por DCVs têm como causas principais ataques cardíacos e derrames, e um terço dessas mortes ocorre prematuramente em pessoas com menos de 70 anos de idade.

> - cerca de 17,9 milhões de pessoas morrem de doenças cardiovasculares por ano
>
> - Mortes por DVCs representam cerca de 32% de todas as causas de mortes

- **Cenário Estados Unidos**

Em 2018, o *Center for Disease Control and Prevention (CDC)* classificou as doenças cardíacas como a principal causa de mortalidade nos Estados Unidos e continua a classificá-la como tanto até os dias de hoje. Devido à complexidade e às variações do número crescente de fatores de risco, o uso de técnicas avançadas como *Machine Learning* vem sendo empregadas para auxiliar no combate contra doenças cardíacas e derrames. Segundo o site da *American Heart Association*, de fevereiro de 2018 até o momento, já houve uma redução de 15,1% nas mortes por doenças cardíacas nos Estados Unidos.

> - cerca de 1 ataque cardíaco a cada 40 segundos
>
> - cerca de 805.000 americanos têm 1 ataque cardíaco por ano
>
> - cerca de 47% de todos os americanos têm pelo menos 1 dos 3 principais fatores de risco para doenças cardíacas: pressão alta, colesterol alto e diabetes

Identificar aqueles com maior risco de DCVs e garantir que eles recebam tratamento adequado pode prevenir mortes prematuras. O acesso a medicamentos para doenças não transmissíveis e tecnologias básicas de saúde em todas as unidades básicas de saúde é essencial para garantir que os necessitados recebam tratamento e aconselhamento.

### <a id = "Análise_Exploratória_de_Dados_(AED)"><font color='#005b96'>Análise Exploratória de Dados (AED)</font></a>

A **Análise Exploratória de Dados (AED)** - *Exploratory Data Analysis (EDA)* - é uma área relativamente nova da estatística e foi desenvolvida por um dos maiores estatísticos de todos os tempos, *John W. Tukey*. Seu livro *Exploratory Data Analysis (EDA)*, lançado em 1977, marca o início dessa abordagem que forma não só a base da análise de dados moderna, mas também os fundamentos da *Ciência de Dados*.

É uma etapa crucial para o sucesso de um projeto de *Ciência de Dados*, pois é através da investigação minuciosa dos dados que se cria uma base sólida de entendimento capaz de propiciar uma extração robusta de informações para que determinados problemas sejam resolvidos de maneira realmente concisa.

Para tanto, a AED envolve a aplicação de técnicas estatísticas e visuais para resumir, examinar e interpretar os dados a fim de descobrir padrões, tendências e características relevantes, sendo uma ponte entre a coleta de dados e as etapas subsequentes da ciência de dados, como a criação de modelos mais avançados e a tomada de decisões mais complexas.

#### <font color='#005b96'>describe()</font>

O método ***describe()*** é utilizado para gerar estatísticas descritivas de um *DataFrame* ou *Series*. Ele calcula várias estatísticas resumidas, como média, desvio padrão, mínimo, máximo e quartis, fornecendo uma visão geral das características dos dados. Ele pode ser aplicado tanto em *DataFrames* quanto em *Series*. Quando aplicado a um *DataFrame*, retorna estatísticas descritivas para todas as colunas do *DataFrame. Quando aplicado a uma *Series*, retorna estatísticas descritivas para os valores da *Series*.

<div class="alert alert-block alert-warning">
<b>Importante:</b> Útil para entender de forma rápida a distribuição e as características dos dados, fornecendo uma visão geral das tendências centrais e da dispersão, facilitando um diagnóstico inicial para uma análise exploratória de dados.
</div>

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.describe.html

```python
# usando describe() para obter um resumo estatístico da Series
df_01['Age'].describe()
```

Para dados tipo `int` e `float` (quantitativos), o resultado incluirá:

|statistic|description|
|:--------|:----------|
|*count*|quantidade de registros|
|*mean*|média|
|*std*|desvio padrão|
|*min*|valor mínimo|
|*25%*|1º quartil|
|*50%*|2º quartil (mediana)|
|*75%*|3º quartil|
|*max*|valor máximo|

Para dados tipo `object` (qualitativos/categóricos), o resultado incluirá:

|statistic|description|
|:--------|:----------|
|*unique*|quantidade de valores únicos|
|*top*|valor mais comum|
|*freq*|frequência do valor mais comum|

#### <a id = "Histograma"><font color='#005b96'>Histograma</font></a>

A maneira mais comum de se visualizar a distribuição dos dados de uma variável quantitativa é por meio de um histograma.

O histograma é a representação gráfica de um conjunto de dados previamente tabulado (uma tabela de frequência) por meio de barras, onde os valores distintos são agrupados em intervalos de classes (***bins***) que são marcadas no *eixo x* formando a base da barra e a frequência com que esses valores agrupados ocorrem são marcados no *eixo y* determinando a altura da barra.

O histograma é uma importante ferramenta estatística de análise, pois permite distinguir diversas características como as medidas de tendência central, a dispersão dos dados e a forma como estão distribuídos.

O construção do histograma possui diversas etapas e cálculos, onde alguns parâmetros possuem ainda a possibilidade de serem calculados através de conceitos matemáticos diferentes entre si.

O principal parâmentro a ser ajustado em um histograma é a quantidade de ***bins***.

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.plot.hist.html

```python
# plotando o histograma
df_01['Age'].plot.hist() # bins=10 by defalut
```

Geralmente, a escolha da quantidade de ***bins*** pode parecer arbitrária, mas essa definição irá depender da necessidade de como inspecionar os dados para se obter as respostas desejadas, o que permite que se experimente diferentes valores para o parâmetro ***bins***.

Uma regra muito utilizada para calcular a quantidade de ***bins*** é o resultado do cálculo da raiz quadrada da quantidade de elementos contidos no conjunto de dados, isso é, &radic;*n*

```python
bins_n = df_01['Age'].shape[0]
print(bins_n)
```

```python
bins_n = np.sqrt(bins_n)
print(bins_n)
```

O parâmetro ***bins*** tem como premissa receber um número inteiro.

```python
bins_n = np.round(bins_n, decimals=0)
print(bins_n)
print(type(bins_n))
```

```python
bins_n = bins_n.astype(int)
print(bins_n)
print(type(bins_n))
```

```python
bins_n = np.round(np.sqrt(df_01['Age'].shape[0]), decimals=0).astype(int)
print(bins_n)
```

```python
# plotando o histograma
df_01['Age'].plot.hist(bins=bins_n)
```

Outra regra para calcular a quantidade de ***bins*** poderia ser a quantidade de valores únicos contidos no conjunto de dados.

```python
df_01['Age'].nunique()
```

```python
# plotando o histograma
df_01['Age'].plot.hist(bins=df_raw['Age'].nunique())
```

***Calculando as faixas de valores (bins) de um histograma usando NumPy***

https://numpy.org/doc/stable/reference/generated/numpy.histogram.html

```python
np.histogram(df_01['Age']) # bins=10 by defalut
```

Acima, a função *np.histogram()* retorna dois *arrays*, onde o primeiro contem a quantidade dos valores agrupados em cada ***bin*** e o segundo determina os limites de valores para cada agrupamento.

**IMPORTANTE**: Todos os *bins*, exceto o último, são semi-abertos, isso é, do primeiro *bin* ao penúltimo *bin* se inclui o primeiro valor e se exclui o segundo valor, mas no último *bin* se inclui os dois valores. Exemplo:

#### <a id = "Boxplot"><font color='#005b96'>Boxplot</font></a>

O *boxplot* é uma ferramenta gráfica que representa uma forma rápida de se visualizar a variação dos dados por meio de quartis.
Através das características informativas do boxplot, torna-se possível visualizar medidas descritivas como o mínimo, máximo, primeiro quartil, segundo quartil (mediana) e terceiro quartil, assim como visualizar a dispersão e a distribuição dos dados como os 50% dos valores mais prováveis – os quais contidos no intervalo interquartil – e a variabilidade além do intervalo interquartil, propiciando a identificação da assimetria dos dados e *outliers*.

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.plot.box.html

```python
# plotando o boxplot
df_01['Age'].plot.box()
```
